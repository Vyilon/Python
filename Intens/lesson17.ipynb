```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import linear_model
from sklearn.cluster import KMeans
from sklearn import metrics 
from pandas import DataFrame
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
my_label = LabelEncoder()    
data['LUNG_CANCER']= my_label.fit_transform(data['LUNG_CANCER'])  
data['GENDER']= my_label.fit_transform(data['GENDER'])  
```

```python
data.head(20)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

data_res = data['LUNG_CANCER']
data_in = data.drop('LUNG_CANCER', axis=1)

x_train, x_test, y_train, y_test = train_test_split(data_in, data_res, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
lr = linear_model.LinearRegression()
lr.fit(x_train, y_train)
lr_pred = lr.predict(x_test)
lr_report = classification_report(y_test, lr_pred.round())
lr_acc = accuracy_score(y_test, lr_pred.round())
```

```python
print(lr_report)
print(lr_acc)
```
